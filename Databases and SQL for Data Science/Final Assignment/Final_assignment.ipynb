```python
df = pd.read_sql('SELECT DISTINCT(PRIMARY_TYPE) FROM CRIME where LOCATION_DESCRIPTION = "GAS STATION"', connection)
df.head()
```

```python
df = pd.read_sql('SELECT * FROM CENSUS', connection)
df.head()
```

```python
df = pd.read_sql('SELECT * FROM CENSUS WHERE COMMUNITY_AREA_NAME LIKE "B%"', connection)
df.head()
```

```python
df = pd.read_sql('SELECT * FROM PUBLIC_SCHOOLS', connection)
df.head()
```

```python
df = pd.read_sql('SELECT NAME_OF_SCHOOL FROM PUBLIC_SCHOOLS WHERE (COMMUNITY_AREA_NUMBER BETWEEN 10 AND 15) AND HEALTHY_SCHOOL_CERTIFIED = "Yes"', connection)
df.head()
```

```python
df = pd.read_sql('SELECT AVG(SAFETY_SCORE) FROM PUBLIC_SCHOOLS', connection)
df.head()
```

```python
df = pd.read_sql('SELECT COMMUNITY_AREA_NAME, AVG(COLLEGE_ENROLLMENT) FROM PUBLIC_SCHOOLS GROUP BY COMMUNITY_AREA_NAME ORDER BY AVG(COLLEGE_ENROLLMENT) DESC LIMIT 5', connection)
df.head()
```

```python
df = pd.read_sql('SELECT SAFETY_SCORE, COMMUNITY_AREA_NAME FROM PUBLIC_SCHOOLS WHERE SAFETY_SCORE = (SELECT MIN(SAFETY_SCORE) FROM PUBLIC_SCHOOLS)', connection)
df.head()
```

```python
df = pd.read_sql('SELECT CENSUS.PER_CAPITA_INCOME, PUBLIC_SCHOOLS.COMMUNITY_AREA_NAME FROM PUBLIC_SCHOOLS, CENSUS WHERE PUBLIC_SCHOOLS.COMMUNITY_AREA_NUMBER = CENSUS.COMMUNITY_AREA_NUMBER AND SAFETY_SCORE = 1', connection)
df.head()
```
